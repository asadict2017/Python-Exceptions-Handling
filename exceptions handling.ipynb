```python
#multiple exception bloc
try:
  1/0
  a=b

except ZeroDivisionError as ex:
  print(ex)
except NameError as ex:
  print(ex)
```

```python
#multiple exception bloc with final exception class
try:
  1/0 #ZeroDivisionError
  a=b #NameError
except NameError as ex:
  print(ex)
except Exception as ex:
  print(ex)
```

```python
# try catch finally ---------->try except else finally
#if no error raise and then exept block will execute finally
try:
  num=int(input("enter a number"))
  result=10/num
except ValueError as ex:
  print(ex)
except ZeroDivisionError as ex:
  print(ex)
except Exception as ex:
  print(ex)
else:
  print(result)
finally:  #all cases it must be executed
  print("this is finally block")
```
